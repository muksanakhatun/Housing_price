##  Importing AND displaying the Data
Load the csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
housing = pd.read_csv('housing_price.csv')
housing.sample(1)
```

- imported all the libraried needed
- we use the method ***SAMPLE*** to display the RANDOM sample 1 columns of the dataframe.

```python
def f(row):
    if row['yr_renovated'] == 0 :
        val = 'no'
    else:
        val = 'yes'
    return val
housing['renovated'] = housing.apply(f, axis=1)
```

```python
housing['age']= 2014 - housing['yr_built']
```

```python
housing['price_sqt_living']= round((housing['price']/ housing['sqft_living']) , 2)
```

### Rplacing missing values
#### Finding the correlation between price and other variables

```python
plt.style.use(['default'])
params= {'axes.labelsize': 11,
        'axes.titlesize': 12,
        'font.size': 12,
        'legend.fontsize': 10,
        'xtick.labelsize': 9,
        'ytick.labelsize': 9,
        'figure.facecolor': 'white',
        'font.family':'serif'}
plt.rcParams.update(params)
```

```python
plt.figure(figsize = (12,3))
sns.boxplot( x= 'price' , data= housing, palette="crest")
plt.title('Box and Whisker plot of price' )
plt.show()
```

```python
plt.figure(figsize = (10,3))
sns.boxplot( x= 'price_sqt_living' , data= housing,palette="crest")
plt.title('Box and Whisker plot of price_sqt_living' )
plt.show()
```

**In the diagram of boxplot shows so many outlies. Thats need to be removed from the datasets. Below some of the price of outlier has been replaced with mean value of price.**

```python
median = housing.price.median()
max_threshold = housing.price.quantile(.995)
max_threshold
```

```python
#housing[housing.price >max_threshold]
```

```python
housing["price"] = np.where(housing["price"] > max_threshold, median,housing['price'])
```

```python
median1 = housing.price_sqt_living.median()
max_threshold1 = housing.price_sqt_living.quantile(.9995)
max_threshold1
```

```python
housing["price_sqt_living"] = np.where(housing["price_sqt_living"] > max_threshold1, median1,housing['price_sqt_living'])
```

```python
plt.figure(figsize = (8,3))
sns.boxplot( x= 'price' , data= housing , palette="crest")
plt.title('Box and Whisker plot of price after replacing missing values' )
plt.show()
```
